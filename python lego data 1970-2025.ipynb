```python
import glob
import pandas as pd
import os
```

```python
# Find all CSV files in the target directory
file_path = "C:/Users/mohit/Desktop/lego data/*.csv"
all_files = glob.glob(file_path)
```

```python
# Load into a dictionary
df_dict = {os.path.basename(f): pd.read_csv(f) for f in all_files}
```

```python
# columns which are not common don't hold any info that would be important for analysis, 
# because category of product or theme is more important than height and weight of the product
```

```python
# 'inner' join keeps only columns common to all files
df = pd.concat(df_dict, axis=0, join='inner', ignore_index=True)
```

```python
df = df.drop('pieces', axis=1)
```

```python
df = df.drop('thumbnailURL', axis=1)
df = df.drop('imageURL', axis=1)
```

```python
df = df.drop('bricksetURL', axis=1)
```

```python
# Drops rows only if they have a missing value in 'themeGroup'
df = df.dropna(subset=['themeGroup'])
```
